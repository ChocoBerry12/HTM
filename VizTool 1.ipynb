```python
import numpy as np
from bokeh.plotting import figure, output_notebook, curdoc
from bokeh.models import Spacer, Paragraph, Button, ButtonGroup, TextEditor, Slider, Panel
from bokeh.layouts import row, column, widgetbox
from bokeh.io import show

N = 10
x = np.linspace(0, 4 * np.pi, N)
y = np.sin(x)
options = dict(tools="", toolbar_location=None, plot_height=300, plot_width=300)


# row 1 - SP
button_1_1 = Button(label="SP create")
button_1_2 = Button(label="init")
slider_1_1 = Slider(start=32, end=2048, value=128, step=103, title="SP size")
slider_1_2 = Slider(start=32, end=2048, value=128, step=103, title="potential ratio")
slider_1_3 = Slider(start=0, end=1, value=.5, step=.05, title="connected threshold")
slider_1_4 = Slider(start=0, end=10, value=1, step=1, title="overlap threshold")

# row 2 - Encoder , SP , data feeding
slider_2_1 = Slider(start=0, end=1, value=.5, step=.05, title="Encoder sparsity")
slider_2_2 = Slider(start=0, end=1, value=.5, step=.05, title="noise")
button_2_1 = Button(label="ang", width=50)
button_2_2 = Button(label="stop", width=50)
button_2_3 = Button(label="one by one", width=50)
button_2_4 = Button(label="continue", width=50)
button_2_5 = Button(label="learn on/off", width=50)
button_2_6 = Button(label="recording", width=50)
button_2_7 = Button(label="")
editbox_2_1 = TextEditor()

# row 3 - Encoder, SP plot
p_3_1 = figure(title="simulation", width=1200)
p_3_1.line(x, y)

# row 4 - recording

# row 5 - status graph
p_5_1 = figure(title="active duty", width=400)
p_5_2 = figure(title="overlap duty", width=400)
p_5_3 = figure(title="sparsity", width=400)

#----------------------------------
# layout 


row_1 = row(column(slider_1_1, slider_1_3), 
            column(slider_1_2, slider_1_4),
            button_1_1,
            button_1_2)

row_2 = row(column(slider_2_1, slider_2_2),
            row(button_2_1, button_2_2, button_2_3),
            row(button_2_4, button_2_5, button_2_6))

row_3 = row(p_3_1)

#row_4 = row()

row_5 = row(p_5_1, p_5_2, p_5_3, height=400)

#row_1 = row([Spacer(width=400), widgets1], sizing_mode=MODE)
#row_2 = row([Paragraph(text=" Encoder "), widgets2, Paragraph(text=" SP ")])
#row_3 = row([p2, p3])
#row_4 = row(p4)
#row_5 = row([p5, p6, p7])


layout = column([row_1, row_2, row_3, row_5])

curdoc().add_root(layout)
```
